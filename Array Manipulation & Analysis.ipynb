(1). You sweet and innocent folks are enrolled as BSC-AI students for a 3 year degree program. You are supposed to study a total of 10 subjects in each particular year. Create a 3D array to represent/store student marks in the form of Data where Shape: (Year, Student, Subject).

# (A).Initialize using np.arange() and reshape().

```python
import numpy as np

years = 3
students = 16   Array Manipulation & Analysis
subjects = 10
total_elements = years * students * subjects # 480 elements


marks = (np.arange(1, total_elements + 1) % 100).reshape((years, students, subjects))

print(f"Array Shape: {marks.shape} (Years, Students, Subjects)\n")
```

# (B).Compute:

(a).Year-wise average :-

```python
year_wise_avg = np.mean(marks, axis=(1, 2))
print("--- Year-wise Average ---")
for i, avg in enumerate(year_wise_avg):
    print(f"Year {i+1} Average: {avg:.2f}")
```

(b).Student-wise total :-

```python
student_wise_total = np.sum(marks, axis=2)
print("\n--- Student-wise Total (I-year)---")
for a, sum in enumerate(student_wise_total):
  print(f"Year {a+1} total: {sum}")
```

(c).Subject-wise highest score :-

```python
subject_wise_highest = np.max(marks, axis=(0, 1))
print("\n--- Subject-wise Highest Score ---")
for i, score in enumerate(subject_wise_highest):
    print(f"Subject {i+1} Highest Score: {score}")
```

# (C). Replace all marks below 40 with 40 (Updating operation).

```python
marks[marks < 40] = 40

print("\n--- Grace Marks Applied! ---")
print(f"Are there any marks below 40 now? {np.any(marks < 40)}")
```

# (D). Sort marks subject-wise.

```python
marks_sorted = np.sort(marks, axis=2)

print("\n--- Marks Sorted Subject-wise (Showing Year 1, Student 1) ---")
print(marks_sorted[0,0])
```

(2). Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target. Assume that each input has only and exactly one solution.

```python
def two_sum_brute(nums, target):
  n = len(nums)
  for i in range(n):
    for j in range(i+1, n):
      if nums[i] + nums[j] == target:
        return [i,j]

two_sum_brute([3, 4, 11, 15, 6], 9)
```
